```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Loading data from the rmsd.xvg file, ignoring lines starting with "@" or "#"
start_nm,d_start = np.loadtxt("density_PVA_1.xvg", comments=["@", "#"], unpack=True)
final_nm,d_final = np.loadtxt("density_PVA_24.xvg", comments=["@", "#"], unpack=True)
d_start = d_start/100
d_final = d_final / 100
```

```python
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(111)

# Plotar ambas as curvas no mesmo eixo
l_plot = ax.plot(start_nm, d_start, color="black", linestyle="-", label="Start")
m_plot = ax.plot(final_nm, d_final, color="blue", linestyle="-", label="End")

# Configurações
ax.set_xlabel("Average coordinate (nm)")
ax.set_ylabel("Density (g/cm$^3$)")
ax.legend()
ax.set_title("Density profile: start vs end")

# Salvar e mostrar
plt.savefig("density_matrix.png", format="png", dpi=300)
plt.show()
```
